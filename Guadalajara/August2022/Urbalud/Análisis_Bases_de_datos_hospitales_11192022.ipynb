```python
def classify(value, breaks):
    for i in range(1, len(breaks)):
        if value < breaks[i]:
            return i
    return len(breaks) - 1

#Determine optimal number of Classes
def goodness_of_variance_fit(array, classes):
    # get the break poin

    # do the actual classification
    classified = np.array([classify(i, classes) for i in array])

    # max value of zones
    maxz = max(classified)

    # nested list of zone indices
    zone_indices = [[idx for idx, val in enumerate(classified) if zone + 1 == val] for zone in range(maxz)]

    # sum of squared deviations from array mean
    sdam = np.sum((array - array.mean()) ** 2)

    # sorted polygon stats
    array_sort = [np.array([array[index] for index in zone]) for zone in zone_indices]

    # sum of squared deviations of class means
    sdcm = sum([np.sum((classified - classified.mean()) ** 2) for classified in array_sort])

    # goodness of variance fit
    gvf = (sdam - sdcm) / sdam

    return gvf

#breaks for ranges 
def breaksForEachColumn (column_name, desc):
  gvf = 0.0
  nclasses = 2
  while gvf < .8:
    breaksTotalHospitals = jenkspy.jenks_breaks(dfCleanByAGEBJal[column_name], n_classes=nclasses)
    arrayOfValuesHospitals = (dfCleanByAGEBJal[column_name].values)
    gvf = goodness_of_variance_fit(arrayOfValuesHospitals, breaksTotalHospitals)
    nclasses += 1

  """print(f"GVF {desc}: {gvf}")
  print(f"Rangos {desc}: {breaksTotalHospitals}")
  print(f"Longitud: {len(breaksTotalHospitals)}")"""
  listOfIntervals = []
  for i, v in enumerate(breaksTotalHospitals):
    if i == 0:
      listOfIntervals.append((v, breaksTotalHospitals[i+1]))
    elif i == (len(breaksTotalHospitals)-1):
      k= 0
    else:
      k = int(v)+1
      listOfIntervals.append((k, breaksTotalHospitals[i+1]))

  return listOfIntervals


print("**          RANGOS PARA ONE HOT ENCODING           **")
print("")
print("")
aOfIntervalsHospitals = breaksForEachColumn("total_hospitales", "Total Hospitales")
print(f"ARRAY INTERVALOS HOSPITALES: {aOfIntervalsHospitals}")
print(""*25)
aOfIntervalsConsultories = breaksForEachColumn("total_consultorios", "Total Consultorios")
print(f"ARRAY INTERVALOS CONSULTORIOS: {aOfIntervalsConsultories}")
print(""*25)
aOfIntervalsDoctors = breaksForEachColumn("total_medicos", "Total Medicos")
print(f"ARRAY INTERVALOS MEDICOS: {aOfIntervalsDoctors}")
print(""*25)
aOfIntervalsEnferemeros =  breaksForEachColumn("total_enfermeros", "Total Enfermeros")
print(f"ARRAY INTERVALOS ENFERMEROS: {aOfIntervalsEnferemeros}")
print(""*25)
aOfIntervalsPracticantes = breaksForEachColumn("total_practicantes_medicos", "Total Practicantes Med")
print(f"ARRAY INTERVALOS PRACTICANTES MEDICOS: {aOfIntervalsPracticantes}")
print(""*25)
aOfIntervalsPobTot = breaksForEachColumn("POBTOT", "Población Total")
print(f"ARRAY INTERVALOS POBTOT: {aOfIntervalsPobTot}")
print(""*25)
aOfIntervalsPobMas = breaksForEachColumn("POBMAS", "Población Masculina Total")
print(f"ARRAY INTERVALOS POBMAS: {aOfIntervalsPobMas}")
print(""*25)
aOfIntervalsPobFem = breaksForEachColumn("POBFEM", "Población Femenina Total")
print(f"ARRAY INTERVALOS POBFEM: {aOfIntervalsPobFem}")
print(""*25)
aOfIntervalsPsinder = breaksForEachColumn("PSINDER", "Población Sin der Total")
print(f"ARRAY INTERVALOS PSINDER: {aOfIntervalsPsinder}")
print(""*25)
aOfIntervalsPderss = breaksForEachColumn("PDER_SS", "Población Sin Seg Total")
print(f"ARRAY INTERVALOS PDER_SS: {aOfIntervalsPderss}")
print(""*25)
aOfIntervalsVivtot = breaksForEachColumn("VIVTOT", "Viviendas Total")
print(f"ARRAY INTERVALOS VIVTOT: {aOfIntervalsVivtot}")
#print(arrayOfValuesHospitals)  
```
