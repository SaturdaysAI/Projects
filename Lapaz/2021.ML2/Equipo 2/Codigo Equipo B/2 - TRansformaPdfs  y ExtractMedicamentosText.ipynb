```
# Almacenara los Regex no encontrados
no_encontrados = []

# Esta esta la estructura que se plantea obtener con los regex
AllMeds = pd.DataFrame( columns=['Grupo', 'Medicamento', 'Generalidades', 'Efectos adversos', 'Contraindicaciones'])

# Para cada uno de los grupos de medicamentos
for i in range(len(grupos)):
    print(f'Grupo: {grupos[i]}')

    grupoMed = pd.DataFrame( columns=['Grupo', 'Medicamento', 'Generalidades', 'Efectos adversos', 'Contraindicaciones'])
    #grupoMed

    # Extraemos la columna que contiene las expresiones regulares a buscar
    regexs = pd.read_excel('/content/drive/MyDrive/Proyecto/Datos/Grupos/Regex.xlsx', sheet_name=grupos[i], header=2, usecols=['Regex'])
    #regexs

    # Abrimos el archivo correspondiente al Grupo, de este archivo se extraen las expresiones regulares de los medicamentos
    with open(f'/content/drive/MyDrive/Proyecto/Datos/Grupos/{grupos[i]}.txt','r') as f:
        # print(f.name)
        read_file = f.read()

    # Mostramos parte de su contenido
    # read_file[:500]

    # Cantidad de Expresiones a encontrar
    # regexs.size
    print('Empieza Regex')

    for j in range(regexs.size):
        print(j,end='; ')
        
        # TRansformamos las regex, para intentar que coincidan al maximo
        rex = regexs.Regex[j].upper()
        match = re.search(rex, read_file.upper()) # Creamos la estructura de busqueda de la regex

        # Se se encuenta la expresion, entonces
        if match:
            # Extrae los grupos delimitados por la regex, estos grupos seran los datos que necesitamos
            medicamento = [f'{grupos[i].upper()}'] + list(match.group(1,4,8,11))
            
            # Y los pasamos al DataFrame
            grupoMed = grupoMed.append(pd.DataFrame([medicamento], columns=['Grupo', 'Medicamento', 'Generalidades', 'Efectos adversos', 'Contraindicaciones']), ignore_index=True)
            #print(medicamento)
        else:
            # Si no se encuentra, hacemos el registro para despues extraerla manualmente
            print("pattern not found", rex)
            no_encontrados.append(grupos[i] + '--' +rex)
    
    print()

    # Almacenamos los datos extraidos
    AllMeds = AllMeds.append( grupoMed, ignore_index=True)
    AllMeds.to_csv('/content/drive/MyDrive/Proyecto/Datos/Grupos/Medicamntos.csv')
```
