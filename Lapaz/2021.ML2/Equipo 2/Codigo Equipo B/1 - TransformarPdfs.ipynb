# 1. Extracción de Datos

Extracción de datos del PDF:

**2020_COFEPRIS_SS_Listado_de_medicamentos_de_Referencia_**

Utilizamos _camelot_, la cual es una librería que permite extraer tablas de datos que estan dentro de un archivo PDF. Esta libreria permite la facíl extracción conservando la mayor cantidad de datos posibles, asi como su estructura.
[Camelot: PDF Table Extraction for Humans](https://camelot-py.readthedocs.io/en/master/)

```
# Importamos Camelot
import camelot
```

```
# Extraemos las tablas del archivo pdf
tables = camelot.read_pdf('/content/drive/MyDrive/Proyecto/Datos/2020_COFEPRIS_SS_Listado_de_medicamentos_de_Referencia_.pdf',pages='all')
```
