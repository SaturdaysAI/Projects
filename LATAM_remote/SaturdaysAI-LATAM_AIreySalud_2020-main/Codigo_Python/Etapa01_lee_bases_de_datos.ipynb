## Revisión de la base horaria con PM2.5 para seleccionar estaciones con suficiencia en 2019

```
# En caso de requerir volver a cargar la PM25_2015_2020
# from google.colab import drive
# drive.mount('/content/drive', force_remount=True)
# import pandas as pd
# import numpy as np
df_PM25=pd.read_csv('/content/drive/My Drive/Colab Notebooks/base_PM25_2015_2020.csv')
#OJO el formato de date es %d-%m-%Y %H:%M con horas de 1 a 24, pero Python no reconoce horas de 1 a 24, las requiere de 0 a 23
#separa la fecha de la hora por el espacio en blanco ' '
df[['Date','Hour']]=df.date.str.split(' ', n = 1, expand = True)
df.Date=pd.to_datetime(df.Date) #formato fecha
df.Hour=df.Hour.str.split(':', n = 1, expand = True)[0] #Dejamos la hora sin minutos
df.Hour=df.Hour.astype(int)-1 #convertimos Hour de string a entero y escalamos de  0 a 23 horas
print(df.info())
df.tail(3)
#generamos un formato datetime en date
df.date=df.Date.astype(str)+" "+df.Hour.astype(str)
#cambiar formato a datetime
df.date = pd.to_datetime(df.date, format="%Y-%m-%d  %H")
```

```
df_PM25['year']=df_PM25['date2'].dt.year
df_PM25.tail(3)
```

```
#Revisa en PM2.5 para cada estación en qué años tiene datos
PM25=df_PM25[df_PM25['id_parameter']=="PM2.5"]
PM25.drop(columns=['id_parameter'],inplace=True) #Elimina la columna parametros
PM25=PM25.groupby(['id_station','year']).count()
PM25=PM25['value']
PM25=PM25.reset_index()
PM25=PM25[PM25['value']>0]
PM25
```

```
#seleccion de estaciones que miden en 2019 con al menos 75% de datos
sitios=PM25[(PM25['year']==2019)&(PM25['value']>=6570)]['id_station']
```

```
df_PM25=df_PM25[df_PM25['id_station'].isin(sitios)]
df_PM25.tail(3)
```

```
#Respalda la base que analizaremos
df_PM25.to_csv('/content/drive/My Drive/Colab Notebooks/base_analisis_2015_2020.csv', index=False)
```
