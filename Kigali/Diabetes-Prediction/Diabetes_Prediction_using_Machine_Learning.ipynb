# II. Data cleaning
======================

### As we've seen on our dataset, we don't have missing values but we have zero values which is not possible in some columns while we're diagnosing our patients. so let's turn them(zero values) into nan values a bit. 

## Task09. Now, Handle missing values or nan values with mean for the sake of having clean data

```python
df_copy = df.copy(deep=True)
```

```python
df_copy.columns
```

```python
df_copy[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI']] = df_copy[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI']].replace(0, np.nan)
```

```python
df_copy
```

```python
df_copy.isnull()
```

```python
df_copy.isnull().sum()
```

```python
df['Glucose'] = df['Glucose'].replace(0, df['Glucose'].mean())
df['BloodPressure'] = df['BloodPressure'].replace(0, df['BloodPressure'].mean())
df['SkinThickness'] = df['SkinThickness'].replace(0, df['SkinThickness'].mean())
df['Insulin'] = df['Insulin'].replace(0, df['Insulin'].mean())
df['BMI'] = df['BMI'].replace(0, df['BMI'].mean())
```

## Task15. Import modules those are necessary in ML Pipeline(Classification) using Scikit-learn

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.pipeline import Pipeline
```

```python
pipeline_lr = Pipeline([("Scalar1", StandardScaler()), ("lr_classifier", LogisticRegression())])
pipeline_knn = Pipeline([("Scalar2", StandardScaler()), ("knn_classifier", KNeighborsClassifier())])
pipeline_svc = Pipeline([("Scalar3", StandardScaler()), ("svc_classifier", SVC())])

pipeline_dt = Pipeline([("dt_classifier", DecisionTreeClassifier(max_depth = 2))])
pipeline_rf = Pipeline([("rf_classifier", RandomForestClassifier(max_depth=3))])
pipeline_gbc = Pipeline([("gbc_classifier", GradientBoostingClassifier(max_depth = 4))])
```

```python
pipelines = [pipeline_lr,
            pipeline_knn,
            pipeline_svc,
            pipeline_dt,
            pipeline_rf,
            pipeline_gbc]
```

```python
pipelines
```

## Task16. Train our available dataset

```python
for pipe in pipelines:
    pipe.fit(X_train, Y_train)
```

## Task17. We choose accuracy as metric to find our best model

```python
pipe_dict = {0:"LinearRegression",
             1:"K-NearestNeighbors",
             2:"SupportVectorClassifier",
             3:"DecisionTree",
             4:"RandomForest",
             5:"GradientBoostingClassifier"}
pipe_dict
```

```python
for i, model in enumerate(pipelines):
    print("{} Test Accuracy:{}".format(pipe_dict[i], model.score(X_train, Y_train)*100))
```
