2.4 Implementación de función para definir tipo de cuerpo

```python
#definir tipo de cuerpo
def get_tipo_cuerpo(medidas,categoria,genero):
    tipoCuerpo = 0
    if genero == 'Femenino':
        #inicia averiguar tipo de cuerpo mujer
        #obtengo medidas de formulario
        hombros=medidas[0]
        busto=medidas[1]
        cintura=medidas[2]
        cadera=medidas[3]
        ombligo=medidas[4]
        cadera=medidas[5]
        #obtengo relación entre medidas
        calc1=hombros/busto
        calc2=cadera/busto
        calc3=cintura/hombros
        calc4=cintura/busto
        calc5=cadera/hombros
        calc6=cadera/busto
        calc7=ombligo/hombros
        calc8=ombligo/busto
        #obtengo mayor entre hombros,busto,cadera
        maximos = [hombros,busto,cadera]
        i = 0 
        posMax=0 
        valorMx=0
        for md in maximos:
            if(md > valorMx):
                valorMx = md
                posMax=i
            i=i+1
        max1 = maximos[posMax]
        i = 0 
        max2 = 0
        max3 = 0
        for md in maximos:
            if(i != posMax):
                if(max2 == 0):
                    max2 = md
                elif(max3 == 0):
                    max3 = md
            i=i+1
        #obtengo mayor y menor entre hombros y caderas
        mayor = 0 
        menor = 0
        if(hombros>cadera):
            mayor = hombros
            menor = cadera
        else:
            menor = hombros
            mayor = cadera
        #defino variables para tipo de cuerpo Rectangulo
        esRectangulo = 0
        bustoRectangulo = 0 
        entreSiRectangulo = 0
        t = max1*0.95
        if(calc3>=0.75 or calc4>=0.75):
            bustoRectangulo = 1
        if(max2>t and max3>t):
            entreSiRectangulo = 1
        if(bustoRectangulo and entreSiRectangulo):
            esRectangulo = 1
        #condiciones tipo de cuerpo mujer
        if(calc1>=1.05 or calc2>1.05):
            tipoCuerpo = 1
            #print('Tu tipo de cuerpo es: Triángulo invertido')
            return tipoCuerpo
        elif(calc5>=1.05 or calc6>=1.05):
            tipoCuerpo = 5
            #print('Tu tipo de cuerpo es: Pera')
            return tipoCuerpo
        elif(calc7>=1.05 or calc8>=1.05):
            tipoCuerpo = 4
            #print('Tu tipo de cuerpo es: Manzana')
            return tipoCuerpo
        if(calc3<0.85 or calc4<0.85):
            bustoReloj = 0
            #definir tipo de cuerpo Reloj de arena
            esReloj = 0 
            if(calc3<0.75 or calc4<0.75):
                bustoReloj = 1
            entreSiRelog = 0
            t = mayor*0.95
            if(menor>t):
                entreSiRelog = 1
            if(bustoReloj and entreSiRelog):
                esReloj = 1
            if(esReloj):
                tipoCuerpo = 2
                #print('Tu tipo de cuerpo es: Relój de arena')
                return tipoCuerpo
            else:
                tipoCuerpo = 5
                #print('Tu tipo de cuerpo es: Pera')
                return tipoCuerpo
        else:
            tipoCuerpo = 3
            #print('Tu tipo de cuerpo es: Rectángulo')
            return tipoCuerpo
        #fin averiguar tipo de cuerpo mujer    
    else:
        #inicia averiguar tipo de cuerpo hombre
        #obtengo medidas de formulario
        hombros=medidas[0]
        ombligo=medidas[1]
        cadera=medidas[2]
        #obtengo relación entre medidas
        calc1=hombros/ombligo
        calc2=cadera/ombligo
        calc3=cadera/hombros
        calc4=cadera/ombligo
        calc5=ombligo/hombros
        calc6=ombligo/cadera
        #condiciones tipo de cuerpo
        if(calc1>=1.05 or calc2>1.05):
            tipoCuerpo = 6
            #print('Tu tipo de cuerpo es: V')
            return tipoCuerpo
        elif(calc5>=1.05 or calc6>=1.05):
            tipoCuerpo = 7
            #print('Tu tipo de cuerpo es: O')
            return tipoCuerpo
        elif(calc3>=1.05 or calc4>=1.05):
            tipoCuerpo = 9
            #print('Tu tipo de cuerpo es: A')
            return tipoCuerpo
        else:
            tipoCuerpo = 8
            #print('Tu tipo de cuerpo es: H')
            return tipoCuerpo
        #inicia averiguar tipo de cuerpo hombre
   
```
