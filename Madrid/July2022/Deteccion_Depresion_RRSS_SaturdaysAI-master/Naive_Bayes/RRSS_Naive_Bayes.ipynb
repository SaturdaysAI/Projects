```python
import pandas as pd
import numpy as np
import re
import time
import emot
from sklearn import datasets
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, roc_curve, auc
import spacy
```

```python
a = time.time()
b = time.time()

#leo el texto eliminando duplicados
data = pd.read_csv('D:\IASaturdays\Proyecto\\limpio.csv')

#renombro columnas y elimino las Id
data = data.drop('Unnamed: 0', axis = 1)
#data = data.sample(25000)
```

```python
#Definimos nuestra x y nuestra y

X = data['texto']
y = data['depresion']
del data
```

```python
#definimos el pipeline

from sklearn.feature_extraction.text import TfidfVectorizer

pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB())
])
```

```python
#primero creamos el vectorizador y el tfidf

tfidf_vectorizer = TfidfVectorizer()

#aplicamos count y tfidf y lo ponemos en un dataframe 
X_tfidf_df = pd.DataFrame(tfidf_vectorizer.fit_transform(X).toarray(), 
                          columns = tfidf_vectorizer.get_feature_names_out())

X_tfidf_df
```

```python
#dividimos en test y train

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 27)
del X, y
```
