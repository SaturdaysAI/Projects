# CCAA SPAIN DATA
Source: https://github.com/datadista/datasets/tree/master/COVID%2019

From May (~18-05-2020) **Datadista** was updating only the national dataset and not the data by autonomous community due to a change in the data collection methodology by the Ministry of Health. 
"The previous series is not comparable with the current one and the Carlos III Health Institute, the body in charge of publishing the historical series, has paralyzed the publication of data until they can correct it".

**Old historical series**: from 21-02-2020 to 18-05-2020

Update: 03-07-2020
The new historical series is now available in three data formats in the Datadista repository. They will keep the datasets from the previous historical series to allow comparison and analysis.

**New historical series**: from 31-01-2020 to 28-06-2020 (weekly update)

Source info **Instituto de Salud Carlos III**: https://cnecovid.isciii.es/covid19/#documentaci%C3%B3n-y-datos

The analysis of the epidemic begins when a cumulative incidence of 5 COVID-19 cases / 100,000 inhabitants is reached in each Autonomous Community and that initial day of analysis is assigned to all the provinces of the respective Autonomous Community.

In order to avoid inconsistencies in the Rt estimates, in the seven-day windows in which there is a majority of “0 cases”, the analysis is completed at the point of the last case + 7 days or at the end of the **epidemiological week of analysis** (whichever is earlier in time).

````
#Este código de R no funciona aquí, pero os lo dejo por si le quereis echar un vistazo:
#DATAFRAME HASTA EL 18-05-2020:
```{r}
library(tidyverse)
#CASOS
ccaa_casos_wider = read.csv("../datadista/ccaa_covid19_casos.csv") #actualizado hasta 21-05-2020
ccaa_casos_wider = ccaa_casos_wider[, 1:90] #selección hasta el 18-05-2020
casos_longer = pivot_longer(ccaa_casos_wider, cols = starts_with("X"), names_to = "fecha", values_to = "total")

casos_longer$fecha = gsub("X", "", casos_longer$fecha)
casos_longer$fecha = as.Date(gsub("[.]", "-", casos_longer$fecha))
casos_longer["tipo"] = "casos"


#HOSPITALIZADOS
ccaa_hospitalizados_wider = read.csv("../datadista/ccaa_covid19_hospitalizados.csv")
ccaa_hospitalizados_wider = ccaa_hospitalizados_wider[,1:90]
hospitalizados_longer = pivot_longer(ccaa_hospitalizados_wider, cols = starts_with("X"), names_to = "fecha", values_to = "total")

hospitalizados_longer$fecha = gsub("X", "", hospitalizados_longer$fecha)
hospitalizados_longer$fecha = as.Date(gsub("[.]", "-", hospitalizados_longer$fecha))
str(hospitalizados_longer)
hospitalizados_longer["tipo"] = "hospitalizados"


#FALLECIDOS
ccaa_fallecidos_wider = read.csv("../datadista/ccaa_covid19_fallecidos.csv")
ccaa_fallecidos_wider = ccaa_fallecidos_wider[,1:78]
dates = c("X2020.02.21", "X2020.02.22", "X2020.02.23", "X2020.02.24", "X2020.02.25", "X2020.02.26", "X2020.02.27", "X2020.02.28", "X2020.02.29","X2020.03.01", "X2020.03.02", "X2020.03.03")
ccaa_fallecidos_wider[dates] <- 0
ccaa_fallecidos_wider = ccaa_fallecidos_wider[, c(1:2, 79:90, 3:78)] #Ordenando las fechas

fallecidos_longer = pivot_longer(ccaa_fallecidos_wider, cols = starts_with("X"), names_to = "fecha", values_to = "total")

fallecidos_longer$fecha = gsub("X", "", fallecidos_longer$fecha)
fallecidos_longer$fecha = as.Date(gsub("[.]", "-", fallecidos_longer$fecha))
str(fallecidos_longer)
fallecidos_longer["tipo"] = "fallecidos"


#UCI
ccaa_uci_wider = read.csv("../datadista/ccaa_covid19_uci.csv")
ccaa_uci_wider = ccaa_uci_wider[,1:90]
uci_longer = pivot_longer(ccaa_uci_wider, cols = starts_with("X"), names_to = "fecha", values_to = "total")

uci_longer$fecha = gsub("X", "", uci_longer$fecha)
uci_longer$fecha = as.Date(gsub("[.]", "-", uci_longer$fecha))
str(uci_longer)
uci_longer["tipo"] = "uci"



#ALTAS
ccaa_altas_wider = read.csv("../datadista/ccaa_covid19_altas.csv")
ccaa_altas_wider = ccaa_altas_wider[, -3]
dates2 = c("X2020.02.21", "X2020.02.22", "X2020.02.23", "X2020.02.24", "X2020.02.25", "X2020.02.26", "X2020.02.27", "X2020.02.28", "X2020.02.29","X2020.03.01", "X2020.03.02", "X2020.03.03", "X2020.03.04", "X2020.03.05", "X2020.03.06", "X2020.03.07", "X2020.03.08")
ccaa_altas_wider[dates2] <- 0
ccaa_altas_wider = ccaa_altas_wider[, c(1:2, 74:90, 3:73)]


altas_longer = pivot_longer(ccaa_altas_wider, cols = starts_with("X"), names_to = "fecha", values_to = "total")

altas_longer$fecha = gsub("X", "", altas_longer$fecha)
altas_longer$fecha = as.Date(gsub("[.]", "-", altas_longer$fecha))
str(altas_longer)
altas_longer["tipo"] = "altas"


#UNIÓN DE TODAS LAS TABLAS
DatosCCAA = rbind(casos_longer, hospitalizados_longer, fallecidos_longer, uci_longer, altas_longer)
str(DatosCCAA)


#DATOS DE POBLACIÓN DE 2019 OBTENIDOS DEL INE (https://www.ine.es/jaxiT3/Datos.htm?t=2853#!tabs-tabla)
poblacion = read_tsv("../PoblacionCCAA.csv",locale = locale(grouping_mark = ".")) #para quitar los puntos que aparecen en el total de población que delimitan los miles
names(poblacion)[1] <- "Comunidades y Ciudades Autonomas"
poblacion = poblacion %>% 
  filter(`Comunidades y Ciudades Autonomas`!="Total",Sexo=="Total",Periodo==2019) %>%
  transmute(cod_ine=str_sub(`Comunidades y Ciudades Autonomas`,1,2),
    Poblacion_2019=Total,CCAA=`Comunidades y Ciudades Autonomas`)
str(poblacion)

poblacion$cod_ine = as.numeric(gsub("(?<![0-9])0+", "", poblacion$cod_ine, perl = TRUE)) #para quitar los ceros delante de los números del código ine


#UNIÓN POBLACIÓN Y DATOS CCAA
DatosCCAA_completo = poblacion %>% select(-CCAA) %>%
  inner_join(DatosCCAA,by="cod_ine")
DatosCCAA_completo = DatosCCAA_completo[, c(4, 1, 3, 5, 6, 2)]
str(DatosCCAA_completo)
#quitamos acentos y ñ:
DatosCCAA_completo$CCAA = as.factor(DatosCCAA_completo$CCAA)
levels(DatosCCAA_completo$CCAA) <- c("Andalucia", "Aragon", "Asturias", "Baleares",
                                     "C. Valenciana", "Canarias", "Cantabria", 
                                     "Castilla La Mancha", "Castilla y Leon", "Catalunya",
                                     "Ceuta", "Extremadura", "Galicia", "La Rioja", "Madrid",
                                     "Melilla", "Murcia", "Navarra", "Pais Vasco")
levels(DatosCCAA_completo$CCAA)


#GUARDAMOS EL DATASET
#Como objeto R
save(DatosCCAA_completo,
     file="../DatosCCAA_completo.Robj")

#Como archivo csv
write_csv(DatosCCAA_completo,
          path="../DatosCCAA_completo.csv")

#CAMBIO EL FORMATO A WIDER PARA QUE SE PUEDAN INCORPORAR MEJOR LOS DATOS DE TEMPERATURAS
datos = read.csv("../DatosCCAA_completo.csv")
datos_wider = pivot_wider(datos, names_from = "tipo", values_from = "total")
str(datos_wider)
#Guardamos como archivo csv
#Como archivo csv
write_csv(datos_wider,
          path="../DatosCCAA_completo_wider.csv")

```
````
