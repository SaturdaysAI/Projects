```
# mostramos datos del dataset donde el NO2 supera una determinada cantidad umbral
# se crea subset con estos registros

umbral_alto = 65
print ('Número de valores que superan el umbral es', len(df[df['NO2']>umbral_alto]), 'de', len(df), ' valores \n\n')
print (df[df['NO2']>umbral_alto])
dfcont = df[df['NO2']>umbral_alto]
```

```
# describe el subset

dfcont.describe().T
```

```
# Histograma de velocidades el viento de valores altos de NO2

plt.title('Velocidades')
plt.hist(dfcont['Veloc'], bins=60, alpha=1, edgecolor = 'black',  linewidth=1)
plt.grid(True)
plt.show()
plt.clf()
```

```
# Histograma de direcciones del viento de registros con valores altos de NO2 (>umbral)

plt.title('Dirección')
plt.hist(dfcont['Direc'], bins=60, alpha=1, edgecolor = 'black',  linewidth=1)
plt.grid(True)
plt.show()
plt.clf()
```

```
# mostramos datos del dataset donde el NO2 está por debajo de una determinada cantidad umbral
# se crea subset con estos registros

umbral_bajo = 3
print ('Número de valores que está por debajo del umbral es', len(df[df['NO2']<umbral_bajo]), 'de', len(df), ' valores \n\n')
print (df[df['NO2']<umbral_bajo])
dfnocont = df[df['NO2']<umbral_bajo]
```

```
# describe el dataset de valores por debajo de umbral

dfnocont.describe().T
```

```
# Histograma de velocidades del viento de valores bajos de NO2

plt.title('Velocidades')
plt.hist(dfnocont['Veloc'], bins=60, alpha=1, edgecolor = 'black',  linewidth=1)
plt.grid(True)
plt.show()
plt.clf()
```

```
# Histograma de direcciones del viento de valores bajos de NO2

plt.title('Dirección')
plt.hist(dfnocont['Direc'], bins=60, alpha=1, edgecolor = 'black',  linewidth=1)
plt.grid(True)
plt.show()
plt.clf()
```

```
# Histograma combinado de Direcciones del viento

plt.title('Dirección')
plt.hist(dfcont['Direc'], bins=60, alpha=1, edgecolor = 'black',  linewidth=1, color = "red")
plt.hist(dfnocont['Direc'], bins=60, alpha=1, edgecolor = 'black',  linewidth=1,color = "green")
plt.grid(True)
plt.show()
plt.clf()
```

```
# Histograma combinado de Velocidades del viento

plt.title('Velocidad')
plt.hist(dfcont['Veloc'], bins=60, range=[0,14], alpha=1, edgecolor = 'black',  linewidth=1, color = "red")
plt.hist(dfnocont['Veloc'], bins=60,range=[0,14], alpha=1, edgecolor = 'black',  linewidth=1,color = "green")
plt.grid(True)
plt.show()
plt.clf()
```

```
# Columnas del dataset

names = df.columns.tolist()
print (names)

# saca los histogramas de todas las variables excepto fecha 
for name in names: 
  if name != "FECHA":
    maximo = df[name].max()
    plt.title(name)
    plt.hist(df[name], bins=60, range=[0,maximo], alpha=1, edgecolor = 'blue',  linewidth=1, color = "blue")
    plt.grid(True)
    plt.show()
    plt.clf()
```

```
# Columnas del dataset

names = df.columns.tolist()
print (names)

# saca los histogramas combinados de todas las columnas excepto fecha para valores altos de NO2 y bajos
for name in names: 
  if name != "FECHA":
    valor_maximo_cont = dfcont[name].max()
    valor_maximo_nocont = dfnocont[name].max()
    if valor_maximo_cont > valor_maximo_nocont : 
      maximo = valor_maximo_cont
    else:
      maximo = valor_maximo_nocont
    plt.title(name)
    plt.hist(dfcont[name], bins=60, range=[0,maximo], alpha=0.7, edgecolor = 'black',  linewidth=1, color = "red")
    plt.hist(dfnocont[name], bins=60,range=[0,maximo], alpha=0.7, edgecolor = 'black',  linewidth=1,color = "green")
    plt.grid(True)
    plt.show()
    plt.clf()
```
