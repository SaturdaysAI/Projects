```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import PowerTransformer
```

```python
data = pd.read_csv('../../processed-data/tidy_socialmedia_data.csv', index_col=0)
```

```python
data.info()
```

```python
data['Date']= pd.to_datetime(data['Date'])
```

```python
dailyMean = data.groupby(["Participant", "Date"], as_index=False)["LowMood"].mean().to_csv('test_RM-ANOVA.csv')
```

```python
test_RMANOVA_long = pd.read_csv('test_RM-ANOVA_long.csv')
test_RMANOVA_long
```

```python
test_RMANOVA_wide = test_RMANOVA_long.pivot(index='Participant', columns='Day')['LowMood']
test_RMANOVA_wide = test_RMANOVA_wide.reset_index(level=['Participant'])
test_RMANOVA_wide
```

```python
data_clustering_mean = pd.read_csv('../../processed-data/cluster_socialmedia_data.csv')
```

```python
data_clustering_mean
```

```python
cluster_ref = data_clustering_mean.Sintomas_Cluster.tolist()
```

```python
test_RMANOVA_wide['Cluster'] = cluster_ref
```

```python
test_RMANOVA_wide
```

```python
test_RMANOVA_wide.to_csv('../../processed-data/test_RMANOVA_wide.csv')
```

```python
test_RMANOVA_long_cluster = pd.melt(test_RMANOVA_wide, id_vars =['Participant', 'Cluster']).sort_values(by=['Participant', 'Day'])
test_RMANOVA_long_cluster.to_csv('../../processed-data/test_RMANOVA_long_cluster.csv')
```
