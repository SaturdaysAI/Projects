## Identificación del número óptimo de clúster 
Identificamos el número optimo de clusters en el conjunto data_clustering_mean (Síntomas de depresión) utilizando el __método de la distorsión o método del codo:__

```python
# Función que generar la gráfica del método Elbow que recibe un conjunto de datos y un título del gráfico:

def plot_elbow(x, figure_name, max_k = 10, n_init = 10):
    inertia = []

    for k in range(1, max_k):
        kmeans = KMeans(n_clusters = k, n_init = n_init).fit(x)
        inertia.append(kmeans.inertia_)

    plot(range(1, max_k), inertia, 'bx-')
    xlabel('k')
    ylabel(u'Dispersión')
    title(figure_name)
    plt.annotate("La dispersión empieza a reducirse",xy = (3.2,75000),xytext = (4,90000),
                 arrowprops = dict(facecolor = "green",shrink = 5, headlength = 5))
```

```python
plot_elbow(data_clustering_mean, 'Elbow: Número de clusters óptimo')
```

Observamos que a partir de __3 k__ el valor promedio de la dispersión se empieza a reducir más lentamente. Sin embargo, para asegurarnos vamos a utilizar otro metodo para definir el número óptimo de clúster llamado __Silhouette.__

```python
# Función que generar la gráfica del método Silhouette que recibe un conjunto de datos y un título del gráfico:

from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist

def plot_silueta(blobs, figure_name, max_k = 10, n_init = 10):
    silueta_avgs = []
    
    for k in range(2, max_k):
        kmean = KMeans(n_clusters = k, n_init = n_init).fit(blobs)
        silueta_avgs.append(silhouette_score(blobs, kmean.labels_))
        
    plot(range(2, max_k), silueta_avgs)
    title(figure_name)
```

```python
plot_silueta(data_clustering_mean, 'silhouette: Número de clusters óptimo')
```

Con este método comprobamos que que el 3 puede ser un buen valor para probar nuestro cluster analysis. Sin embargo, se ha decidio en el equipo agrupar el conjunto de datos también en 5 clústers.

## Implementación del algoritmo y visualización de los grupos en un gráfico 3D

## Clustering 5 K

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams["figure.figsize"] = (15,10)
plt.style.use('ggplot')

from sklearn.cluster import KMeans

color_map = array(['b','g','r','k', 'y'])

## Aplicar el método de k-means para obtener los centroides  

kmeans = KMeans(n_clusters = 5, random_state = 1000,).fit(data_clustering_mean) 
centroides = kmeans.cluster_centers_ # localizar la posición de los centroides

classes = kmeans.predict(data_clustering_mean) # Obtener a que clúster pertenece cada participante
etiquetas = kmeans.labels_
inercia = kmeans.inertia_
    
# Gráfico

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data_clustering_mean.iloc[:, 0], data_clustering_mean.iloc[:, 1], data_clustering_mean.iloc[:, 3],
           color = color_map[classes], s = 60)
ax.scatter(centroides[:, 0], centroides[:, 1], centroides[:, 3],  marker = '*', color = color_map, s = 1000)

title('Clustering K-means: FaceMood 5k')
```

```python
print("La posición de los centroides es:\n", centroides, "\n")
```

```python
data_clustering_mean['Sintomas_Cluster'] = etiquetas
listado_participante_cluster = data_clustering_mean['Sintomas_Cluster'].sort_values(ascending = True)
```

### Listado de participantes y clúster correspondiente

```python
print("Listado de participantes y clúster correspondiente:\n")
print(listado_participante_cluster)
```

```python
data_clustering_mean.head(10).round()
```

### Ditribución de participantes en cada cluster

```python
sns.set_style('whitegrid')
g=sns.factorplot(x="Sintomas_Cluster", data=data_clustering_mean, kind='count')
ax = plt.gca()
y_max = data_clustering_mean['Sintomas_Cluster'].value_counts().max()
ax.set_ylim([0, (y_max+4)])
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()),
            fontsize=12, color='black', ha='center', va='bottom')
plt.show()
```

```python
print("Participantes/cluster")
print(data_clustering_mean.Sintomas_Cluster.value_counts())
```

```python
data_clustering_mean = pd.concat([data_mean[['PSMU', 'News', 'ASMU']], data_clustering_mean], axis = 1, sort = False)
```

### Subconjuntos de datos para elaborar los histogramas solapados y poder observar graficamente las diferencias entre cada grupo.

```python
RRSS_PSMU = RRSS.groupby('Sintomas_Cluster3')['PSMU']
RRSS_News = RRSS.groupby('Sintomas_Cluster3')['News']
RRSS_ASMU = RRSS.groupby('Sintomas_Cluster3')['ASMU']
RRSS_Fatigue = RRSS.groupby('Sintomas_Cluster3')['Fatigue']
RRSS_LowMood = RRSS.groupby('Sintomas_Cluster3')['LowMood']
RRSS_Loneliness = RRSS.groupby('Sintomas_Cluster3')['Loneliness']
RRSS_Concentrat = RRSS.groupby('Sintomas_Cluster3')['Concentrat']
RRSS_LossOfInt = RRSS.groupby('Sintomas_Cluster3')['LossOfInt']
RRSS_Inferior = RRSS.groupby('Sintomas_Cluster3')['Inferior']
RRSS_Hopeless = RRSS.groupby('Sintomas_Cluster3')['Hopeless']
RRSS_Stress = RRSS.groupby('Sintomas_Cluster3')['Stress']
```

```python
# Función para imprimir histogramas solapados  

def printOverlapHistograms(data, titulo, ejex):
    sns.set_style("white") 
    sns.set_context("paper", font_scale = 4, rc = {"lines.linewidth": 2.5})
    data.plot.hist(alpha = 0.85, width = 2.5)
    sns.despine()
    plt.legend()
    plt.title(titulo)
    plt.xlabel(ejex)
    plt.xlim([0,80])
```

### Histogramas solapados variables RRSS

En el caso de las variables relacionadas con RRSS no encontramos grandes diferencias entre clusters.

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
fig = plt.figure(figsize = (50, 25))
gs = GridSpec(nrows = 2, ncols = 2)
fig.add_subplot(gs[0, 0])
printOverlapHistograms(RRSS_PSMU, "PSMU BY CLUSTER", "PSMU level")
fig.add_subplot(gs[0, 1])
printOverlapHistograms(RRSS_News, "News BY CLUSTER", "News level")
fig.add_subplot(gs[1, 0])
printOverlapHistograms(RRSS_ASMU, "ASMU BY CLUSTER", "ASMU level")
plt.show()
```

### Histogramas solapados síntomas de depresión

Se observa que los 3 clusters son diferentes en lo que concierne a los síntomas relacionados con la depresión, siendo el Cluster 1 el que contiene las mayores puntuaciones medias en relación a los 8 síntomas del estudio, por lo que se podría afirmar que es el grupo con mayor propensión a presentar un posible cuadro de depresión.

```python
fig = plt.figure(figsize = (70, 70))
gs = GridSpec(nrows = 4, ncols = 2)
fig.add_subplot(gs[0, 0])
printOverlapHistograms(RRSS_LowMood, "LowMood BY CLUSTER", "LowMood level")
fig.add_subplot(gs[0, 1])
printOverlapHistograms(RRSS_LossOfInt, "LossOfInt BY CLUSTER", "LossOfInt level")
fig.add_subplot(gs[1, 0])
printOverlapHistograms(RRSS_Hopeless, "Hopeless BY CLUSTER", "Hopeless level")
fig.add_subplot(gs[1, 1])
printOverlapHistograms(RRSS_Stress, "Stress BY CLUSTER", "Stress level")
fig.add_subplot(gs[2, 0])
printOverlapHistograms(RRSS_Fatigue, "Fatigue BY CLUSTER", "Fatigue level")
fig.add_subplot(gs[2, 1])
printOverlapHistograms(RRSS_Loneliness, "Loneliness BY CLUSTER", "Loneliness level")
fig.add_subplot(gs[3, 0])
printOverlapHistograms(RRSS_Concentrat, "Concentrat BY CLUSTER", "Concentrat level")
fig.add_subplot(gs[3, 1])
printOverlapHistograms(RRSS_Inferior, "Inferior BY CLUSTER", "Inferior level")
plt.show()
```
